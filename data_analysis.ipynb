```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 20})
```

```python
df_classrooms = pd.read_csv("./datasets/clean/CaracterizaçãoDasSalasSmall.csv", sep=";", header=0)
print(len(df_classrooms))
df_classrooms.head(5)
```

```python
df_classes = pd.read_csv("./datasets/clean/HorarioDeExemploSmall.csv", sep=";", header=0)
df_classes["Início - Fim"] = df_classes.apply(lambda x: x["Início"] + " - " + x["Fim"], axis=1)
print(len(df_classes))
df_classes.head(5)
```

```python
def plot_histogram(data, title):
  plt.figure(figsize=(24, 12))
  sns.histplot(data, kde=True)
  plt.title(title, fontdict={"fontsize": 24})
  plt.xlabel("Values")
  plt.ylabel("Frequency")
  plt.xticks(rotation=90, ha="right")

  plt.tight_layout()
  plt.show()

def plot_bars(data, title):
  plt.figure(figsize=(24, 12))
  sns.barplot(x=data.index, y=data.values)
  plt.title(title, fontdict={"fontsize": 24})
  plt.xlabel("Values")
  plt.ylabel("Frequency")
  plt.xticks(rotation=90, ha="right")

  plt.tight_layout()
  plt.show()
```

```python
plot_histogram(df_classes["Curso"], "Curso")
plot_histogram(df_classes["Inscritos no turno"], "Inscritos no turno")
plot_histogram(df_classrooms["Capacidade Normal"], "Capacidade das salas")
plot_histogram(df_classes["Dia da Semana"], "Dia da Semana")
plot_histogram(df_classes["Início - Fim"].sort_values(), "Início - Fim")
plot_histogram(df_classes["Dia"], "Dia")
plot_histogram(df_classes["Sala da aula"], "Sala da aula")
plot_histogram(df_classes["Características da sala pedida para a aula"], "Características da sala pedida para a aula")
plot_bars((df_classrooms[df_classrooms.columns[5:]] == "X").sum(), "Características das salas")
```
