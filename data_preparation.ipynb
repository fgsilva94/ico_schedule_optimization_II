```python
import pandas as pd
```

```python
df_classrooms = pd.read_csv("./datasets/original/CaracterizaçãoDasSalas.csv", sep=";", header=0)
print(len(df_classrooms))
df_classrooms.head(5)
```

```python
df_classes = pd.read_csv("./datasets/original/HorarioDeExemplo.csv", sep=";", header=0)
df_classes.drop(df_classes[df_classes["Características da sala pedida para a aula"] == "Não necessita de sala"].index, inplace=True)
df_classes.drop(df_classes[df_classes["Dia"].isna()].index, inplace=True)
df_classes.drop(df_classes[df_classes["Inscritos no turno"].isna()].index, inplace=True)
df_classes.drop(df_classes[df_classes["Inscritos no turno"] < 5].index, inplace=True)
df_classes.drop(df_classes[df_classes["Sala da aula"].isna()].index, inplace=True)
df_classes.drop(df_classes[df_classes["Sala da aula"] == "0S08"].index, inplace=True)
df_classes.loc[df_classes["Características da sala pedida para a aula"] == "Sala/anfiteatro aulas", "Características da sala pedida para a aula"] = "Sala de Aulas normal"
df_classes = df_classes[df_classes["Curso"].apply(lambda x: x.startswith(("L", "M")))]
df_classes.reset_index(drop=True, inplace=True)

print(len(df_classes))
df_classes.head(5)
```

```python
df_classes_small = df_classes[df_classes["Curso"].apply(lambda x: x.startswith(("M")))]
df_classes_small.reset_index(drop=True, inplace=True)

print(len(df_classes_small))
df_classes_small.head(5)
```

```python
used_classrooms = df_classes["Sala da aula"].unique()
df_classrooms = df_classrooms[df_classrooms["Nome sala"].isin(used_classrooms)]
df_classrooms.reset_index(drop=True, inplace=True)

print(len(df_classrooms))
df_classrooms.head(5)
```

```python
used_classrooms_small = df_classes_small["Sala da aula"].unique()
df_classrooms_small = df_classrooms[df_classrooms["Nome sala"].isin(used_classrooms_small)]
df_classrooms_small.reset_index(drop=True, inplace=True)

print(len(df_classrooms_small))
df_classrooms_small.head(5)
```

```python
df_classes.to_csv("./datasets/clean/HorarioDeExemplo.csv", sep=";", header=True, index=False)
df_classrooms.to_csv("./datasets/clean/CaracterizaçãoDasSalas.csv", sep=";", header=True, index=False)

df_classes_small.to_csv("./datasets/clean/HorarioDeExemploSmall.csv", sep=";", header=True, index=False)
df_classrooms_small.to_csv("./datasets/clean/CaracterizaçãoDasSalasSmall.csv", sep=";", header=True, index=False)
```
